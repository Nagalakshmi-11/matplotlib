```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(start=-10,stop=10,num=400)
```

```python
m=2
c=3
y=m*x+c 
```

LINEAR EQUATION

```python
plt.plot(x,y,label='y=2x+3')
plt.title('plot of the Linear Equation')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(0,10)
plt.ylim(0,20)
plt.grid(True)
plt.legend()
plt.savefig('linear-equation.png')
plt.show()            
```

 QUADRATIC EQUATION

```python
a=-1
b=-4
c=4
x=np.linspace(-1,7,400)
y=a*x**2+b*x+c
plt.plot(x,y,label='y=x^2-4x+4')
plt.title('Plot of the Quadratic Equation')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)         
plt .show()         
         
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
fig=plt.figure(figsize=(8,6))
ax=fig.add_axes([0.3,0.3,0.85,0.85])
ax.plot(x,y,label='sin(x)',color='purple')
ax.set_title('plot of sin(x)')
ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('Amplitude')
plt.show()
```

BAR GRAPHS

```python
categories=['Category A','Category B','Category C',]
values=[10,15,7]
plt.bar(categories,values)
plt.title('Bar Graph')
plt.xlabel('Categories')
plt.ylabel('Values')
 
plt.show()
```

```python
categories=['Category A','Category B','Category C',]
values=[10,15,7]
plt.bar(categories,values,color=['red','green','blue'],edgecolor='yellow')
plt.title(' Colored Bar Graph')
plt.xlabel('Categories')
plt.ylabel('Values')
 
plt.show()
```

```python
categories=['Category A','Category B','Category C',]
values1=[10,15,7]
values2=[5,7,3]

plt.bar(categories,values1, label='part1')
plt.bar(categories,values2,  label='part2')
plt.title('Stacked Bar Graph')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.legend()
 
plt.show()
```
